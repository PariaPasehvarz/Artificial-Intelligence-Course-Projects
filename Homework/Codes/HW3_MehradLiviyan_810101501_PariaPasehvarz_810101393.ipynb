# Advantages of adding more layers to deep neural network

**Advantages:**

1. Hierarchy of Features: Deep networks can learn a hierarchy of features. Lower layers learn simple features like edges and textures, while deeper layers combine these simple features to learn more complex ones. This is particularly useful for complex datasets where higher-level features are derived from lower-level ones.

2. Model Capacity: Deeper networks have more parameters and hence a greater capacity to learn. They can model complex functions that shallow networks cannot. This is beneficial when dealing with complex datasets where the underlying patterns are not linearly separable or are high-dimensional.

3. Representation Learning: Deep networks are good at representation learning, which means they can transform raw data into a more suitable internal representation or feature vector. This can make it easier for subsequent layers or classifiers to make accurate predictions.

4. End-to-End Learning: Deep networks can learn to extract features and make predictions in an end-to-end manner, which can lead to better performance on complex tasks compared to traditional machine learning models that require manual feature engineering.

# Problems

While increasing the depth of a neural network can often improve its ability to learn complex patterns in the data, it does not always lead to optimum performance. Here are a few reasons why:


## **1 - Overfitting:**
Deeper networks have more parameters, which can make them more prone to overfitting. Overfitting occurs when the model learns the training data too well, including its noise and outliers, and performs poorly on unseen data.

## **2 - Vanishing/Exploding Gradients:**
As the network becomes deeper, it becomes more difficult for the network to learn because of the vanishing or exploding gradients problem. This is a problem where the gradients, which are used to update the weights, become either too small or too large. This can slow down training or cause it to fail altogether.

## **3 - Computational Complexity:**
Deeper networks require more computational resources and time to train. Depending on the available resources, it might not be feasible to increase the depth of the network.

## **4 - Diminishing Returns:**
After a certain point, adding more layers might not lead to better performance. The improvement in performance with additional layers may become negligible after a certain depth.

# **Which datasets have better performance when we add more layers to the deep neural network model?**

Increasing the depth of a neural network can improve proficiency on a variety of datasets, particularly those with complex hierarchical patterns.

1. Image Classification Datasets: Datasets like ImageNet1, CIFAR, and MNIST2 often benefit from deeper networks. Deeper networks like VGG16, ResNet, and DenseNet have achieved state-of-the-art results on these datasets. These networks have hundreds of layers and have been successful in learning hierarchical features in images.[1][2]

2. Complex models:

* Complex Patterns: When the data has complex hierarchical
patterns, deeper networks can learn these patterns better than shallow networks. This is often the case with high-resolution images, where complex features need to be learned from raw pixel data.

* Large Datasets: Deeper networks have more parameters and hence have a greater capacity to learn. So, when you have a large amount of data, deeper networks can be beneficial as they can learn more complex functions.

* Advanced Architectures: Some advanced network architectures like Residual Networks (ResNets) and Dense Convolutional Networks (DenseNets) are specifically designed to benefit from increased depth. These architectures use techniques like skip connections to mitigate issues like vanishing gradients that typically come with increased depth.

# **Which datasets have worse performance when we add more layers to the deep neural network model?**

Increasing the depth of a neural network does not always lead to improved performance. In fact, on some datasets, it can even decrease proficiency due to various reasons such as overfitting, vanishing/exploding gradients, and increased computational complexity.

1. Small Datasets: If the dataset is small, a deep neural network with many parameters might overfit the training data, leading to poor generalization on unseen data.[1]

2. Simple Tasks: For relatively simple tasks or datasets where the underlying patterns can be captured by shallow networks, increasing the depth might not lead to significant improvements and could even hurt performance.[1]

3. Local Connectivity: Datasets that benefit from local connectivity, such as image datasets where nearby pixels are more correlated, might not benefit from very deep networks. Convolutional Neural Networks (CNNs) with local connectivity often perform well on these tasks with fewer layers.[2]

4. Stability Training: In some cases, increasing the depth of the network can lead to instability in training, which can decrease proficiency.[3]

## **Fashion MNIST**

We used Fashion MNIST as an image classification dataset.

 The Fashion MNIST dataset consists of 70,000 grayscale images(training set of 60,000 and a test set of 10,000) of clothing items.


The images are 28x28 pixels in size and represent 10 different classes of clothing items, including T-shirts/tops, trousers, pullovers, dresses, coats, sandals, shirts, sneakers, bags, and ankle boots.

###**Import necessary libraries**

```python
import keras
from keras.datasets import fashion_mnist
from keras.layers import Conv2D , MaxPooling2D , Flatten
import keras
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder , StandardScaler
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from keras.layers import Dense , BatchNormalization , Dropout
from keras.losses import  SparseCategoricalCrossentropy
import matplotlib.pyplot as plt
```

###**Load Data**

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

###**See some features of csv_file**

```python
print(x_train.shape, y_train.shape)
print(x_train)
```

###**Preproccesing**

```python
x_train= x_train.reshape(x_train.shape[0],28,28,1)
x_test=  x_test.reshape(x_test.shape[0],28,28,1)
input_shape=(28,28,1)
x_train= x_train.astype('float32')
x_test= x_test.astype('float32')
x_train /= 255
x_test /=255
```

###**Split our data to train , validation , test**

```python
x_train , x_val , y_train , y_val = train_test_split(x_train , y_train , train_size=0.90 , random_state=42)
print(y_val.shape,x_val.shape)
```

### Build shallow model :

```python
numberOfFilters = 32
numberOfClasses = 10
shallowModel = keras.Sequential([
    Conv2D(numberOfFilters , kernel_size=(3 ,3) , activation='relu' , input_shape=input_shape),
    MaxPooling2D(pool_size = (2 , 2)),
    Flatten(),
    Dense(256 , activation = 'relu'),
    Dense(numberOfClasses , activation = 'softmax'),
])
```

### Compile model :

```python
shallowModel.compile(optimizer=keras.optimizers.Adam() , loss=SparseCategoricalCrossentropy() , metrics=['accuracy'])
```

### Fit our model :

```python
history = shallowModel.fit(x_train , y_train , batch_size=2000 , epochs=100 , validation_data=(x_val , y_val))
```

### Evaluating our model :

```python
testLoss , testAcc = shallowModel.evaluate(x_test , y_test)
trainLoss , trainAcc = shallowModel.evaluate(x_train , y_train)
valLoss , valAcc = shallowModel.evaluate(x_val , y_val)
```

```python
print('test Loss : ' , testLoss , 'test Accuarcy : ' , testAcc)
print('train Loss : ' , trainLoss , 'train Accuarcy : ' , trainAcc)
print('val Loss : ' , valLoss , 'val Accuarcy : ' , valAcc)
```

### Loss :

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.legend()
plt.grid(True)
```

### Accuarcy :

```python
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='test')
plt.xlabel('Epoch')
plt.ylabel('accuracy')
plt.legend()
plt.grid(True)
```

### Build deep model :

```python
numberOfFilters = 32
numberOfClasses = 10
deepModel = keras.Sequential([
    Conv2D(numberOfFilters , kernel_size=(3 ,3) , activation='relu' , input_shape=input_shape),
    MaxPooling2D(pool_size = (2 , 2)),
    Conv2D(numberOfFilters , kernel_size=(3 ,3) , activation='relu' ),
    MaxPooling2D(pool_size = (2 , 2)),
    Conv2D(numberOfFilters , kernel_size=(3 ,3) , activation='relu' ),
    MaxPooling2D(pool_size = (2 , 2)),
    Dropout(0.2) ,
    Flatten(),
    Dense(256 , activation = 'relu'),
    BatchNormalization(),
    Dropout(0.2),
    Dense(200 , activation = 'relu'),
    BatchNormalization(),
    Dropout(0.2) ,
    Dense(256 , activation = 'relu'),
    BatchNormalization(),
    Dense(256 , activation = 'relu'),
    BatchNormalization(),
    Dropout(0.2),
    Dense(numberOfClasses , activation = 'softmax'),
])
```

### Compile model :

```python
deepModel.compile(optimizer=keras.optimizers.Adam() , loss=SparseCategoricalCrossentropy() , metrics=['accuracy'])
```

### Fit our model :

```python
history = deepModel.fit(x_train , y_train , batch_size=2000 , epochs=100 , validation_data=(x_val , y_val))
```

### Evaluating our model :

```python
testLoss , testAcc = deepModel.evaluate(x_test , y_test)
trainLoss , trainAcc = deepModel.evaluate(x_train , y_train)
valLoss , valAcc = deepModel.evaluate(x_val , y_val)
```

```python
print('test Loss : ' , testLoss , 'test Accuarcy : ' , testAcc)
print('train Loss : ' , trainLoss , 'train Accuarcy : ' , trainAcc)
print('val Loss : ' , valLoss , 'val Accuarcy : ' , valAcc)
```
